```python
import pandas as pd
```

```python
df = pd.read_csv('./mimic-iv-ecg-diagnostic-electrocardiogram-matched-subset-1.0/machine_measurements.csv')
df.head()
```

```python
df['report_1'].astype(str) + df['report_2'].astype(str)
```

```python
df['reports'] = df['report_1'].astype(str)
for i in range(2, 18):
    df['reports'] = df['reports'] + df[f"report_{i}"].astype(str)
df.head()
```

```python
df = df[["study_id", "reports"]].set_index('study_id')
```

```python
df["st_elevation"] = df["reports"].str.contains("elevation")
df.head()
```

```python
df["st_depression"] = df["reports"].str.contains("depression")
df.head()
```

```python
df["t_wave"] = df["reports"].str.contains("T wave change")
df.head()
```

```python
icd_df = pd.read_csv('./mimic-iv-ecg-ext-icd-diagnostic-labels-for-mimic-iv-ecg-1.0.1/records_w_diag_icd10.csv')
```

```python
icd_df = icd_df[['study_id', 'ed_hadm_id', 'hosp_hadm_id', 'all_diag_all']].set_index('study_id')
icd_df.head()
```

```python
df = df.join(icd_df, how='inner', rsuffix='icd')
df.head()
```

```python
Acute_MI = [
    "I210",
    "I211",
    "I212",
    "I213",
    "I214",
    "I219",
    "I220",
    "I221",
    "I228",
    "I229",
    "I230",
    "I231",
    "I232",
    "I233",
    "I234",
    "I235",
    "I236",
    "I238",
]

def isAcuteMI(all_diag_all):
    for icd in eval(all_diag_all):
        for miIcd in Acute_MI:
            if miIcd in icd:
                return True
    return False

df['Acute_MI'] = df['all_diag_all'].map(isAcuteMI)
df.head()
```

```python
df['Acute_MI'].sum()
```

```python
df['hadm_id'] = df['hosp_hadm_id'].fillna(df['ed_hadm_id'])
df = df.drop(['hosp_hadm_id', 'ed_hadm_id'], axis=1)
df.head()
```

```python
discharge_df = pd.read_csv('./mimic-iv-note/note/discharge.csv')
discharge_df.head()
```

```python
discharge_df = discharge_df[['hadm_id', 'text']].set_index('hadm_id')
discharge_df.head()
```

```python
import re

SECTIONS = ['Allergies', 'Chief Complaint', 'Major Surgical or Invasive Procedure', 'History of Present Illness', 'Past Medical History', 'Social History', 'Family History', 'Medications on Admission']
def clean_note(note):
    cleaned = ''
    for heading in SECTIONS:
        content = note.split(heading+':')
        if len(content) == 1: continue
        content = re.split(r'\n[^\n:]+:', content[1])[0] # find next section heading and cut
        cleaned += heading+':' + content + '\n'
    return cleaned

discharge_df['text'] = discharge_df['text'].map(clean_note)
discharge_df.head()
```

```python
df['study_id'] = df.index
df = df.set_index('hadm_id').join(discharge_df, how='inner')
```

```python
lab_df = []
with pd.read_csv('mimic-iv-3.0/hosp/labevents.csv', chunksize=10**6) as reader:
    for chunk in reader:
        chunk = chunk[chunk['hadm_id'].isin(df.index)]
        chunk = chunk[chunk['itemid'].isin([51002, 51003, 52642])]
        lab_df.append(chunk)

lab_df = pd.concat(lab_df)
lab_df.head()
```

```python
df['troponin'] = False
df.loc[lab_df['hadm_id'].unique(), 'troponin'] = True
df.head()
```

```python
df['STEMI'] = df['Acute_MI'] & df['st_elevation']
df['NSTEMI'] = df['Acute_MI'] & ~df['st_elevation']
df = df.drop('reports', axis=1)
df
```

```python
df.to_csv('mimic-acute-mi.csv')
```
